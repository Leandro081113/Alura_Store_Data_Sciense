### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total por tienda
ingreso_tienda_1 = tienda["Precio"].sum()
ingreso_tienda_2 = tienda2["Precio"].sum()
ingreso_tienda_3 = tienda3["Precio"].sum()
ingreso_tienda_4 = tienda4["Precio"].sum()

# Mostrar los ingresos
print("Ingreso total Tienda 1:", ingreso_tienda_1)
print("Ingreso total Tienda 2:", ingreso_tienda_2)
print("Ingreso total Tienda 3:", ingreso_tienda_3)
print("Ingreso total Tienda 4:", ingreso_tienda_4)
```

```python


# Datos
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
labels = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Gráfico de torta
plt.figure(figsize=(6,6))
plt.pie(ingresos, labels=labels, autopct="%1.1f%%", colors=plt.cm.Set3.colors)
plt.title("Distribución de Ingresos Totales por Tienda")
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría - Tienda 1
categorias_t1 = tienda["Categoría del Producto"].value_counts()
print("Tienda 1:")
print(categorias_t1)

# Tienda 2
categorias_t2 = tienda2["Categoría del Producto"].value_counts()
print("\nTienda 2:")
print(categorias_t2)

# Tienda 3
categorias_t3 = tienda3["Categoría del Producto"].value_counts()
print("\nTienda 3:")
print(categorias_t3)

# Tienda 4
categorias_t4 = tienda4["Categoría del Producto"].value_counts()
print("\nTienda 4:")
print(categorias_t4)
```

```python
# Contar categorías
cat_t1 = tienda["Categoría del Producto"].value_counts()
cat_t2 = tienda2["Categoría del Producto"].value_counts()
cat_t3 = tienda3["Categoría del Producto"].value_counts()
cat_t4 = tienda4["Categoría del Producto"].value_counts()

# Crear figura con 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Donut - Tienda 1
axs[0, 0].pie(cat_t1, labels=cat_t1.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
axs[0, 0].set_title("Tienda 1")

# Donut - Tienda 2
axs[0, 1].pie(cat_t2, labels=cat_t2.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
axs[0, 1].set_title("Tienda 2")

# Donut - Tienda 3
axs[1, 0].pie(cat_t3, labels=cat_t3.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
axs[1, 0].set_title("Tienda 3")

# Donut - Tienda 4
axs[1, 1].pie(cat_t4, labels=cat_t4.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
axs[1, 1].set_title("Tienda 4")

# Ajustar diseño
plt.suptitle("Categorías más vendidas por tienda", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda (redondeado a 1 decimal)
promedio_t1 = round(tienda["Calificación"].mean(), 2)
promedio_t2 = round(tienda2["Calificación"].mean(), 2)
promedio_t3 = round(tienda3["Calificación"].mean(), 2)
promedio_t4 = round(tienda4["Calificación"].mean(), 2)

# Mostrar los resultados
print("Promedio de calificación - Tienda 1:", promedio_t1)
print("Promedio de calificación - Tienda 2:", promedio_t2)
print("Promedio de calificación - Tienda 3:", promedio_t3)
print("Promedio de calificación - Tienda 4:", promedio_t4)
```

```python
# Crear DataFrame con los promedios
import pandas as pd
calificaciones = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Promedio Calificación": [promedio_t1, promedio_t2, promedio_t3, promedio_t4]
})

# Gráfico de puntos (línea con marcadores)
plt.figure(figsize=(8,5))
plt.plot(calificaciones["Tienda"], calificaciones["Promedio Calificación"], marker='o', color='purple', linewidth=2)
plt.title("Calificación Promedio por Tienda")
plt.ylim(0, 5)
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Contar productos por tienda
productos_t1 = tienda["Producto"].value_counts()
productos_t2 = tienda2["Producto"].value_counts()
productos_t3 = tienda3["Producto"].value_counts()
productos_t4 = tienda4["Producto"].value_counts()

# Mostrar más y menos vendidos
print("Tienda 1:")
print("Más vendido:", productos_t1.idxmax(), "-", productos_t1.max(), "ventas")
print("Menos vendido:", productos_t1.idxmin(), "-", productos_t1.min(), "ventas\n")

print("Tienda 2:")
print("Más vendido:", productos_t2.idxmax(), "-", productos_t2.max(), "ventas")
print("Menos vendido:", productos_t2.idxmin(), "-", productos_t2.min(), "ventas\n")

print("Tienda 3:")
print("Más vendido:", productos_t3.idxmax(), "-", productos_t3.max(), "ventas")
print("Menos vendido:", productos_t3.idxmin(), "-", productos_t3.min(), "ventas\n")

print("Tienda 4:")
print("Más vendido:", productos_t4.idxmax(), "-", productos_t4.max(), "ventas")
print("Menos vendido:", productos_t4.idxmin(), "-", productos_t4.min(), "ventas\n")
```

```python
# Top 5 más vendidos por tienda
plt.figure(figsize=(12,6))

# Tienda 1
plt.subplot(2, 2, 1)  # 2 filas, 2 columnas, primera posición
productos_t1.head(5).plot(kind="bar", color="lightblue")
plt.title("Top 5 productos más vendidos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

# Tienda 2
plt.subplot(2, 2, 2)  # 2 filas, 2 columnas, segunda posición
productos_t2.head(5).plot(kind="bar", color="lightblue")
plt.title("Top 5 productos más vendidos - Tienda 2")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

# Tienda 3
plt.subplot(2, 2, 3)  # 2 filas, 2 columnas, tercera posición
productos_t3.head(5).plot(kind="bar", color="lightblue")
plt.title("Top 5 productos más vendidos - Tienda 3")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

# Tienda 4
plt.subplot(2, 2, 4)  # 2 filas, 2 columnas, cuarta posición
productos_t4.head(5).plot(kind="bar", color="lightblue")
plt.title("Top 5 productos más vendidos - Tienda 4")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Bottom 5 menos vendidos por tienda
plt.figure(figsize=(12,6))

# Tienda 1
plt.subplot(2, 2, 1)  # 2 filas, 2 columnas, primera posición
productos_t1.tail(5).plot(kind="bar", color="lightcoral")
plt.title("Top 5 productos menos vendidos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

# Tienda 2
plt.subplot(2, 2, 2)  # 2 filas, 2 columnas, segunda posición
productos_t2.tail(5).plot(kind="bar", color="lightcoral")
plt.title("Top 5 productos menos vendidos - Tienda 2")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

# Tienda 3
plt.subplot(2, 2, 3)  # 2 filas, 2 columnas, tercera posición
productos_t3.tail(5).plot(kind="bar", color="lightcoral")
plt.title("Top 5 productos menos vendidos - Tienda 3")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

# Tienda 4
plt.subplot(2, 2, 4)  # 2 filas, 2 columnas, cuarta posición
productos_t4.tail(5).plot(kind="bar", color="lightcoral")
plt.title("Top 5 productos menos vendidos - Tienda 4")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
envio_t1 = round(tienda["Costo de envío"].mean(), 2)
envio_t2 = round(tienda2["Costo de envío"].mean(), 2)
envio_t3 = round(tienda3["Costo de envío"].mean(), 2)
envio_t4 = round(tienda4["Costo de envío"].mean(), 2)

# Mostrar resultados
print("Costo de envío promedio - Tienda 1:", envio_t1)
print("Costo de envío promedio - Tienda 2:", envio_t2)
print("Costo de envío promedio - Tienda 3:", envio_t3)
print("Costo de envío promedio - Tienda 4:", envio_t4)
```

```python
# Crear DataFrame resumen
envios = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo Envío Promedio": [envio_t1, envio_t2, envio_t3, envio_t4]
})

# Graficar
plt.figure(figsize=(8,5))
plt.bar(envios["Tienda"], envios["Costo Envío Promedio"], color="orange")
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo Promedio ($)")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

#**Informe Final de Análisis de Tiendas**


## Introducción:

El objetivo de este análisis es determinar cuál de las cuatro tiendas es la menos conveniente para que el Sr.
Juan continúe operando, basándose en factores clave como ingresos, productos vendidos, calificaciones de
clientes, categorías de productos y costos de envío.

## Desarrollo del análisis:

1. Ingresos Totales:
- La Tienda 4 obtuvo los ingresos más altos.
- La Tienda 1 y la Tienda 3 tuvieron ingresos moderados.
- La Tienda 2 presentó los ingresos más bajos.
2. Productos más y menos vendidos:
- Tienda 4 destacó con varios productos entre los más vendidos.
- Tiendas 2 y 3 tuvieron varios productos con bajas ventas.
3. Calificaciones promedio:
- Tienda 3 tiene la mejor calificación.
- Tienda 4 mantiene una calificación aceptable.
- Tienda 2 mostró menor satisfacción del cliente.
4. Categorías más populares:
- Tienda 4 tiene una buena diversidad de categorías vendidas.
- Tienda 1 se enfoca en tecnología.
- Tienda 2 en hogar y decoración.
5. Costo promedio de envío:
- Tienda 2 tiene el costo más bajo.
- Tienda 4 y Tienda 1 están en rango medio.
- Tienda 3 tiene el costo más alto.




## Conclusión:


Recomendamos que el Sr. Juan se deshaga de la Tienda 2. Esta tienda tiene un bajo rendimiento general en
comparación con las otras, y sus debilidades son notorias en varios aspectos:
- **Bajos ingresos totales**, lo que indica una menor demanda o precios bajos.
- **Calificación más baja** por parte de los clientes, lo que afecta su reputación.
- **Pocos productos vendidos**, con una baja rotación de inventario.
- **Pocas categorías populares** que no destacan frente a las demás tiendas.
- Aunque su **costo de envío es bajo**, las otras tiendas superan este factor con mejor rendimiento.
Vender esta tienda le permitiría al Sr. Juan enfocar sus recursos y esfuerzos en las tiendas con mejor
desempeño.
